```python
import numpy as np
import re

def ReadDiagonally(matrix):
    """Returns a matrix that is readen diagonally"""
    rows, cols = matrix.shape
    newmat = []

    for line in range(1, rows + cols):
        start_col = max(0, line - rows)
        count = min(line, (cols - start_col), rows)
        
        # For this line:
        newmat.append(["".join(matrix[min(rows, line)- i - 1][start_col + i] for i in range(count))])
    return newmat


data = open("input.txt", "r").readlines()
data = np.array([[char for char in string if char != "\n"] for string in data])

Lines = np.array([])

# Horizontal
Lines = np.append(Lines, "".join(np.concatenate(data)))
# Vertical
Lines = np.append(Lines, "".join(np.concatenate(np.rot90(data))))
# First diagonal
Lines = np.append(Lines, "".join(np.concatenate(ReadDiagonally(data))))
# Second diagonal
Lines = np.append(Lines, "".join(np.concatenate(ReadDiagonally(np.rot90(data)))))


for line in Lines:
    highlighted_line = line.replace("XMAS", "\033[91mXMAS\033[0m").replace("SAMX", "\033[91mSAMX\033[0m")
    print(highlighted_line)

count = len(np.concatenate([re.findall(r"(?=XMAS)|(?=SAMX)", i) for i in Lines]))
count
```

```python
import numpy as np

def Search(index, array, cols, rows, word = "XMAS"):
    directions = np.array([
        #X, Y
        [0, 1],  #N 
        [1, 0],  #E
        [0, -1], #S
        [-1, 0], #W
        [1, 1],  #NW
        [1, -1], #SE
        [-1, -1],#SW
        [-1, 1]])  #NW
        
    
    count = 0
    for direction in directions:
        Found = True

        for i in range(len(word)):
            sx, sy = index + i * direction
            # print(sx, sy)

            if sx > cols - 1 or sy > rows - 1 or sx < 0 or sy < 0:
                Found = False
                break
            
            # print(f"Cmp: {word[i]} == {array[sx][sy]}")
            if array[sx][sy] != word[i]:
                Found = False
                break
            
            # If this passes, the word was matched!
        if Found:
            count += 1
        # print(count)
        # print()
    
    # print()

    return count
        


data = open("input.txt", "r").readlines()
data = np.array([[char for char in string if char != "\n"] for string in data])

rows, cols = data.shape

Sum = 0
for x in range(rows):
    for y in range(cols):
        if data[x, y] == "X":
            Sum += Search(np.array([x, y]), data, cols, rows)

print("Part 1: ", Sum)
```

```python
with open("input.txt", 'r') as file:
    lines = file.readlines()
    array = np.array([list(line.rstrip()) for line in lines])

print(array)

# part 2
def check_subarray(subarray):
    check = ["MAS", "SAM"]
    s1 = "".join(subarray.ravel()[[0, 4, 8]])
    s2 = "".join(subarray.ravel()[[2, 4, 6]])
    return (s1 in check) and (s2 in check)

x_mas_count = 0
for i in range(1, array.shape[0]-1):
    for j in range(1, array.shape[0]-1):
        subarray = array[i-1:i+2, j-1:j+2]
        x_mas_count += check_subarray(subarray)

print(f"Total X-MAS count: {x_mas_count}")

```
