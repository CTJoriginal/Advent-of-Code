```python
import numpy as np
import matplotlib.pyplot as plt    

data = open("input.txt").read()

dir = {
    ">": np.array([0, 1]),
    "^": np.array([1, 0]),
    "<": np.array([0, -1]),
    "v": np.array([-1, 0])
}

houses = set()
housesp2 = set()

CurPos = np.array([0, 0])

for i in data:
    houses.add(tuple(CurPos))
    CurPos += dir[i]

CurPos = np.array([0, 0])
CurPosRobot = np.array([0, 0])
robot = True    

for i in data:
    robot = not robot
    if robot: 
        housesp2.add(tuple(CurPosRobot))
        CurPosRobot += dir[i]
    else:
        housesp2.add(tuple(CurPos))
        CurPos += dir[i]
    
print("Part 1:", len(houses))
print("Part 2:", len(housesp2))


x_coords, y_coords = zip(*houses)
plt.scatter(x_coords, y_coords, s=1)
x_coords, y_coords = zip(*housesp2)
plt.scatter(x_coords, y_coords, s=1)
plt.show()
```
